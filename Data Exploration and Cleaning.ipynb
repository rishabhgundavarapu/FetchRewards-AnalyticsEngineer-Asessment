### Reading in data

```python
import pandas as pd
brands = pd.read_json('brands.json',lines=True)
receipts = pd.read_json('receipts.json',lines=True)
users = pd.read_json('users.json',lines=True)
receipts
```

## Lets perform some preprocessing 
### Changing column names


```python
users = users.rename(columns={'_id':'userId'})
brands = brands.rename(columns={"_id": "brandId"})
receipts = receipts.rename(columns={'_id':'receiptId'})
users
```

### Some of the records are nested, lets correct this by getting the original values present


```python
def replace_dict_with_oid_or_date(df):
    for col in df.columns:
        if df[col].apply(lambda x: isinstance(x, dict)).any():
            first_dict = df[col].iloc[0]
            if any(key.startswith('$oid') or key.startswith('$date') for key in first_dict.keys()):
                df[col] = df[col].apply(lambda x: x[list(filter(lambda k: k.startswith('$oid') or k.startswith('$date'), x.keys()))[0]] if isinstance(x, dict) else x)
    return df
users = replace_dict_with_oid_or_date(users)
brands = replace_dict_with_oid_or_date(brands)
receipts = replace_dict_with_oid_or_date(receipts)
users
```

## Creating new tables

```python
## CPG TABLE
def extract_values(collection_dict):
    if isinstance(collection_dict, dict):
        oid = collection_dict['$id']['$oid']
        ref = collection_dict['$ref']
        return pd.Series({'oid': oid, 'ref': ref})
    else:
        return pd.Series({'oid': None, 'ref': None})
cpg = brands['cpg'].apply(extract_values)
brands.drop('cpg',axis=1,inplace=True)
cpg
```

### Retrieving nested information from the receipts file

```python

def explode_dict_lists(df, col_name):
    all_keys = set()
    for record in df[col_name]:
        if isinstance(record, list):
            for item in record:
                if isinstance(item, dict):
                    all_keys.update(item.keys())
    new_df = pd.DataFrame(columns=sorted(all_keys))
    
    for idx, row in df.iterrows():
        record = row[col_name]
        id_value = row['receiptId']
        if isinstance(record, list):
            new_row = pd.Series({key: None for key in all_keys}, index=all_keys)
            for item in record:
                item.update({'receiptId':id_value})
                if isinstance(item, dict):
                    for key, value in item.items():
                        new_row[key] = value
                    new_row['receiptId'] = id_value
                    new_df.loc[len(new_df)] = new_row
        else:
        
            new_row = pd.Series({key: None for key in all_keys}, index=all_keys)
            new_row['receiptId'] = id_value
            new_df.loc[len(new_df)] = new_row

    return new_df
rewardsReceiptItemInfo= explode_dict_lists(receipts,'rewardsReceiptItemList')
receipts.drop(['rewardsReceiptItemList'],axis=1,inplace=True)
rewardsReceiptItemInfo
```

## Answering Business Questions using SQL Queries

### Q1. What are the top 5 brands by receipts scanned for most recent month?

#### There are two ways this question can be interpreted:
#### 1. For the month of the latest date scanned, get the top 5 brands (if the last date was July 3 2021, we would get all records for july 2021)
#### 2. Get the top 5 brands from a period of 1 month from the latest date scanned (for eg if latest date scanned was July 11 2021, this would mean records from June 11 to July 11)

#### I will follow the second approach as it is the common way of how companies store data 

```python
query = """ with most_recent_month_sales as ( SELECT receiptId
FROM receipts
WHERE dateScanned >= ((SELECT MAX(dateScanned) FROM receipts) - (31 * 24 * 60 * 60 * 1000))
)

SELECT brandCode,COUNT(*) as brand_count
FROM receipts rec
INNER JOIN rewardsReceiptItemInfo re ON rec.receiptId = re.receiptId
WHERE rec.receiptId IN (SELECT receiptId FROM most_recent_month_sales)
GROUP BY brandCode
ORDER BY brand_count DESC


  """

result = pysqldf(query)
result
```

#### It looks like the brandcode column is mostly null for the last month period from the receipts file. There are only 3 other brands namely BRAND,MISSION,VIVA
#### This seems to be an error that likely occured during data collection or during data entry

### Q2. How does the ranking of the top 5 brands by receipts scanned for the recent month compare to the ranking for the previous month?

```python
query = """ with most_recent_month_sales as ( SELECT receiptId
FROM receipts
WHERE dateScanned <= (SELECT MAX(dateScanned) FROM receipts) - (31 * 24 * 60 * 60 * 1000)
and dateScanned >= (SELECT MAX(dateScanned) FROM receipts) - (62 * 24 * 60 * 60 * 1000)
)

SELECT brandCode,COUNT(*) as brand_count
FROM receipts rec
INNER JOIN rewardsReceiptItemInfo re ON rec.receiptId = re.receiptId
WHERE rec.receiptId IN (SELECT receiptId FROM most_recent_month_sales)
GROUP BY brandCode
ORDER BY brand_count DESC
LIMIT 6


  """

result = pysqldf(query)
result
```

#### Although there is a lot of missing data in the previous month (None value counts are the highest again), we can see a healthier and denser distribution of brands
#### The TOP 5 brands in the previous month are Ben and Jerry's , Hyvee, Kroger, Borden and Pepsi

### Q3. When considering average spend from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?

```python
query = """
        SELECT distinct(rewardsReceiptStatus)
        FROM receipts
        """

result = pysqldf(query)
result
```

#### Upon simply looking at the data, there is no status such as ACCEPTED for the receipts anywhere, None of the other values match closely in meaning except for FINISHED,
#### We will assume that FINISHED status has the same meaning as ACCEPTED, although it could mean something else entirely too, ideally this issue should be checked for and solved as early as possible 

```python
query = """
        SELECT AVG(totalSpent),rewardsReceiptStatus
        FROM receipts
        WHERE rewardsReceiptStatus IN ('FINISHED','REJECTED')
        GROUP BY rewardsReceiptStatus
        """


# Execute the SQL query
result = pysqldf(query)
result
```

#### Receipts with status of FINISHED/ACCEPTED have a higher spend (80.85) as compared to receipts with a REJECTED status (23.32)

### Q4. When considering total number of items purchased from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?

```python
query = """
        SELECT SUM(purchasedItemCount),rewardsReceiptStatus
        FROM receipts
        WHERE rewardsReceiptStatus IN ('FINISHED','REJECTED')
        GROUP BY rewardsReceiptStatus
        """
result = pysqldf(query)
result
```

#### Receipts with FINISHED status have more total number of items purchased (8184) as compared to receipts with a REJECTED status (173)

### Q5. Which brand has the most spend among users who were created within the past 6 months?

```python
query = """ WITH six_month_user_ids AS (
SELECT distinct(userId)
FROM users
WHERE createdDate >= (SELECT MAX(createdDate) FROM users) - (186 * 24 * 60 * 60 * 1000)
order by createdDate
)

select userId,sum(totalSpent) as tot,rew.brandCode from (select * from six_month_user_ids su inner join receipts re on su.userId = re.userId) as meow
inner join rewardsReceiptItemInfo rew on rew.receiptId = meow.receiptId
group by rew.brandCode
order by tot desc
LIMIT 1


  """

result = pysqldf(query)
result
```

#### Ben and Jerrys has the highest spend across users who created an account in the last six months

### Q6. Which brand has the most transactions among users who were created within the past 6 months?

```python
query = """ WITH six_month_user_ids AS (
SELECT distinct(userId)
FROM users
WHERE createdDate >= (SELECT MAX(createdDate) FROM users) - (186 * 24 * 60 * 60 * 1000)
order by createdDate
)

select count(rew.brandCode) as transactions,rew.brandCode from (select * from six_month_user_ids su inner join receipts re on su.userId = re.userId) as meow
inner join rewardsReceiptItemInfo rew on rew.receiptId = meow.receiptId
group by rew.brandCode
order by transactions

  """

result = pysqldf(query)
result
```

#### HY-VEE has the most transactions among users who created an account in the last six months

### Missing Data

```python
def percentage_missing_data(df):
    missing_data = df.isnull().sum()
    percentage_missing = (missing_data / df.shape[0]) * 100
    return percentage_missing
percentage_missing_data(users)
```

```python
percentage_missing_data(brands)
```

```python
percentage_missing_data(receipts)
```
